```python
import os
import re
import yaml
```

```python
NOTES_DIR = r"Obsidian - Copie"
```

```python
pattern = re.compile(r"^(?P<tag>[^-]+?)\s*-\s*(?P<name>.+)\.md$")
# pattern = re.compile(r"^(?P<tag>p)\s(?P<name>.+)\.md$")


def process_file(filepath):
    filename = os.path.basename(filepath)
    match = pattern.match(filename)
    if not match:
        return  # fichier non conforme

    tag = match.group("tag").strip()
    # tag = "personnage"
    note_name = match.group("name").strip()
    new_filename = f"{note_name}.md"
    new_filepath = os.path.join(os.path.dirname(filepath), new_filename)

    with open(filepath, "r", encoding="utf-8") as f:
        content = f.read()

    if content.startswith("---"):
        # Séparer frontmatter et corps
        parts = content.split("---", 2)
        yaml_block = parts[1]
        body = parts[2]

        data = yaml.safe_load(yaml_block) or {}

        # Normaliser "tags"
        tags = data.get("tags")
        if tags is None:
            tags = []
        elif isinstance(tags, str):
            tags = [tags]
        elif not isinstance(tags, list):
            tags = [str(tags)]

        if tag not in tags:
            tags.append(tag)

        data["tags"] = tags

        # Réécrire le YAML
        new_yaml = yaml.safe_dump(data, sort_keys=False, allow_unicode=True).strip()
        new_content = f"---\n{new_yaml}\n---{body}"

    else:
        # Pas de frontmatter : on crée un nouveau
        yaml_header = f"---\ntags:\n  - {tag}\n---\n\n"
        new_content = yaml_header + content

    # Écrire le nouveau fichier sous le bon nom
    with open(new_filepath, "w", encoding="utf-8") as f:
        f.write(new_content)

    # Supprimer l’ancien fichier si renommage
    if new_filepath != filepath:
        os.remove(filepath)

    print(f"✅ {filename} → {new_filename} (tag ajouté: {tag})")
```

```python
def main():
    for fname in os.listdir(NOTES_DIR):
        if fname.endswith(".md"):
            process_file(os.path.join(NOTES_DIR, fname))

if __name__ == "__main__":
    main()
```
