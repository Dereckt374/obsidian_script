```python
import os
from utils_obsidian import add_yaml_field_from_md
from manage_obsidian_pro import OBS_PRO, OBS_LOS
```

Ce script ajoute un champ yaml & une valeur à tous les fichiers .md contenus dans un repertoire

```python
# -----------------------------
# 🔧 VARIABLES À MODIFIER
# -----------------------------
DIRECTORY = os.path.join(OBS_LOS, "Specs")
FIELD = "field" 
FIELD_VALUE = ""
RECURSIVE = False   # mettre False pour non-récursif
# -----------------------------
```

```python
def main():
    if RECURSIVE:
        for root, _, files in os.walk(DIRECTORY):
            for file in files:
                if file.endswith(".md"):
                    add_yaml_field_from_md(os.path.join(root, file), FIELD, FIELD_VALUE)
    else:
        for file in os.listdir(DIRECTORY):
            if file.endswith(".md"):
                add_yaml_field_from_md(os.path.join(DIRECTORY, file), FIELD, FIELD_VALUE)

if __name__ == "__main__":
    main()
```
