```python
import os
import re
```

```python
def nettoyer_fichier(entree, sortie):
    with open(entree, "r", encoding="utf-8") as f_in, open(sortie, "w", encoding="utf-8") as f_out:
        buffer = ""

        for ligne in f_in:
            ligne = ligne.rstrip("\n")  # on enlève le retour à la ligne pour analyser

            # Vérifie si la ligne se termine par un chiffre, une virgule ou un point
            if re.search(r"[0-9,.;:]$", ligne):
                # On écrit ce qu'on a accumulé + la ligne actuelle + un vrai retour à la ligne
                buffer += ligne
                f_out.write(buffer + "\n")
                buffer = ""
            else:
                # Ligne à fusionner avec la suivante → pas de retour à la ligne
                buffer += ligne + " "

        # Si quelque chose reste dans le buffer à la fin
        if buffer.strip():
            f_out.write(buffer.strip())
```

```python
def reformater_definitions(fichier_entree, fichier_sortie):
    with open(fichier_entree, "r", encoding="utf-8") as f_in, \
         open(fichier_sortie, "w", encoding="utf-8") as f_out:

        for ligne in f_in:
            ligne = ligne.strip()
            if not ligne:
                continue  # ignore les lignes vides

            # On sépare au premier deux-points
            if ":" in ligne:
                terme, definition = ligne.split(":", 1)
                terme = terme.strip()
                definition = definition.strip()

                f_out.write(f"###### {terme}\n")
                f_out.write(f"{definition}\n\n")
            else:
                # Si la ligne n'a pas de ":", on la recopie telle quelle
                f_out.write(ligne + "\n")
```

```python
path_in = r'C:\Users\virgil.mesle\OneDrive - Sirius Space Services\Documents\LOS\obsidian_los\Définitions.md'
path_out = r'C:\Users\virgil.mesle\OneDrive - Sirius Space Services\Documents\LOS\obsidian_los\Définitions_cleaned.md'
```

```python
nettoyer_fichier(path_in, path_out)
```

```python
reformater_definitions(path_in, path_out)
```
